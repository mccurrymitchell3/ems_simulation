```python
## SEVERITY

# 1. For each incident type, output the severity it is most often initially assigned
severity_map_br = {}
for index, row in br.iterrows():
    if row['INITIAL_CALL_TYPE'] not in severity_map_br.keys():
        severity_map_br[row['INITIAL_CALL_TYPE']] = {row['INITIAL_SEVERITY_LEVEL_CODE']: 1}
    else:
        if row['INITIAL_SEVERITY_LEVEL_CODE'] not in severity_map_br[row['INITIAL_CALL_TYPE']]:
            severity_map_br[row['INITIAL_CALL_TYPE']][row['INITIAL_SEVERITY_LEVEL_CODE']] = 1
        else:
            severity_map_br[row['INITIAL_CALL_TYPE']][row['INITIAL_SEVERITY_LEVEL_CODE']] += 1
            
severity_map_br
```

```python
sev_modes_br = {}
for inj in severity_map_br.keys():
    maxOcc = 0
    for j in severity_map_br[inj].keys():
        if severity_map_br[inj][j] > maxOcc:
            mode = j
            maxOcc = severity_map_br[inj][j]
    sev_modes_br[inj] = mode
    
sev_modes_br
```

```python
# 2. For each incident type, output the severity it is most often finally assigned
severity_map_brf = {}
for index, row in br.iterrows():
    if row['FINAL_CALL_TYPE'] not in severity_map_brf.keys():
        severity_map_brf[row['FINAL_CALL_TYPE']] = {row['FINAL_SEVERITY_LEVEL_CODE']: 1}
    else:
        if row['FINAL_SEVERITY_LEVEL_CODE'] not in severity_map_brf[row['FINAL_CALL_TYPE']]:
            severity_map_brf[row['FINAL_CALL_TYPE']][row['FINAL_SEVERITY_LEVEL_CODE']] = 1
        else:
            severity_map_brf[row['FINAL_CALL_TYPE']][row['FINAL_SEVERITY_LEVEL_CODE']] += 1
            
sev_modes_brf = {}
for inj in severity_map_brf.keys():
    maxOcc = 0
    for j in severity_map_brf[inj].keys():
        if severity_map_brf[inj][j] > maxOcc:
            mode = j
            maxOcc = severity_map_brf[inj][j]
    sev_modes_brf[inj] = mode
    
sev_modes_brf
```

```python
# 3. determine if/what they often change to:
diffs = {}
for injtype in sev_modes_brf.keys():
    try:
        if sev_modes_br[injtype] != sev_modes_brf[injtype]:
            diffs[injtype] = (sev_modes_br[injtype], sev_modes_brf[injtype])
    except:
        print(injtype)
            
diffs

```

```python
## CONCLUSION: WE WILL USE THE FINAL SEVERITY CODES TO DETERMINE COMMONALITY, 
## AS THOSE ARE MOST REPRESENTATIVE OF WHEN AMBULANCES WILL BE DISPATCHED
```

```python
# 4. figure out commonality of each severity level
#       use final (since that's what the calls are based on)

type(br['FIRST_ON_SCENE_DATETIME'][0])
type(br['FIRST_ON_SCENE_DATETIME'][18])

sent_notsent = {}
for index, row in br.iterrows():
    if row['FINAL_SEVERITY_LEVEL_CODE'] not in sent_notsent.keys():
        sent_notsent[row['FINAL_SEVERITY_LEVEL_CODE']] = {'sent': 0, 'not sent': 0}
    if type(row['FIRST_ON_SCENE_DATETIME']) == str:
        sent_notsent[row['FINAL_SEVERITY_LEVEL_CODE']]['sent'] += 1
    else:
        sent_notsent[row['FINAL_SEVERITY_LEVEL_CODE']]['not sent'] += 1
        
sent_notsent
        
```
