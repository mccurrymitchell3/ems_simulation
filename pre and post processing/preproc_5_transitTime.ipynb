```python
import pandas as pd
import numpy as np
import math
import datetime as dt
```

```python
br = pd.read_csv("bronx.csv")
```

```python
# determine avg time between incident and dispatch for real data 
calltime = []
for row in br['INCIDENT_DATETIME']:
    idt = row.replace("T", " ")
    calltime.append(dt.datetime.strptime(idt, '%Y-%m-%d %H:%M:%S.%f'))
    
calltime.reverse() #because the data is saved from 12/31/2018-1/1/2015

dispatchtime = []
for row in br['FIRST_ASSIGNMENT_DATETIME']:
    if type(row) == str:
        idt = row.replace("T", " ")
        dispatchtime.append(dt.datetime.strptime(idt, '%Y-%m-%d %H:%M:%S.%f'))
    else:
        dispatchtime.append("del")
dispatchtime.reverse()
```

```python
dispatchtime = []
for row in br['FIRST_ASSIGNMENT_DATETIME']:
    if type(row) == str:
        idt = row.replace("T", " ")
        dispatchtime.append(dt.datetime.strptime(idt, '%Y-%m-%d %H:%M:%S.%f'))
    else:
        dispatchtime.append("del")
dispatchtime.reverse()

closetime = []
for row in br['INCIDENT_CLOSE_DATETIME']:
    if type(row) == str:
        idt = row.replace("T", " ")
        closetime.append(dt.datetime.strptime(idt, '%Y-%m-%d %H:%M:%S.%f'))
    else:
        closetime.append("del")
closetime.reverse()
```

```python
# clean to get rid of nans 
indsToDelete = [i for i, j in enumerate(dispatchtime) if j == 'del']
#len(indstodel)
for i in range(len(indsToDelete)-1,-1, -1):
    del dispatchtime[indsToDelete[i]]
    del closetime[indsToDelete[i]]
    
    
indsToDelete2 = [i for i, j in enumerate(closetime) if j == 'del']
for i in range(len(indsToDelete2)-1,-1, -1):
    del dispatchtime[indsToDelete2[i]]
    del closetime[indsToDelete2[i]]
```

```python
print([i for i, j in enumerate(dispatchtime) if j == 'del'])
print([i for i, j in enumerate(closetime) if j == 'del'])
```

```python
transittimes = [closetime[i] - dispatchtime[i] for i in range(len(closetime))]
avgtransittime = np.mean(transittimes)
avgtransittime.seconds/60
```

```python
transitsecs = [x.seconds for x in transittimes]
np.std(transitsecs)/60
```
